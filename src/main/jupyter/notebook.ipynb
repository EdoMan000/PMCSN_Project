```python
import pandas as pd


def load_csv(file: str) -> pd.DataFrame:
    return pd.read_csv(file)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

def plot_observations(center_name: str, df: pd.DataFrame):
    if not os.path.exists('plots'):
        os.makedirs('plots')
    # Plot the data
    fig = plt.figure(figsize=(10, 6))
    plt.plot(df.index, df["E[Ts]"], marker='o', linestyle='-', color='b')

    # Add title and labels
    plt.title(f'Welch Plot for {center_name}')
    plt.xlabel('Index')
    plt.ylabel('E[Ts]')
    
    # Show the plot
    plt.grid(True)
    plt.savefig(f'plots/{center_name}', dpi=fig.dpi)
    plt.show()

import os

parent = '../../../csvFiles/BASIC_SIMULATION_EXPONENTIAL/observations'
for f in os.listdir(parent):
    df = load_csv(str(os.path.join(parent, f)))
    plot_observations(f.replace('.csv', ''), df)
```
